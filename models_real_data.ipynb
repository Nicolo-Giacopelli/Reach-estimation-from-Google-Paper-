### Simple OLS

```python
store_dem = {}

for (d_group, train), (d_group2, test) in zip(couples_X_Y_TR.items(), couples_X_Y_TE.items()):
    X_train, Y_train = train[0], train[1]    
    X_test, Y_test = test[0], test[1]     # we evaluate on dataset of same demographic group in January

    r2, rmse = [], []

    lr = LinearRegression().fit(X_train, Y_train)
    preds = lr.predict(X_test)

    r2 = r2_score(Y_test, preds)
    rmse = np.sqrt(mean_squared_error(Y_test, preds))
    
    print(f"\n\t \t DEMOGRAPHIC GROUP: {d_group}")
    print(f"R2 score on January's sample: {r2: .5%}")
    print(f"RMSE score on January's sample: {rmse}")
    
    store_dem[d_group] = [r2, rmse]
```

```python
store_dem = {}

spl = 100
tot_chosen = []

alpha_grid = np.concatenate((np.logspace(-4, 0, 150), np.arange(1.1, 1.5, 0.1)))  # 100 observations from 1e-4 until 2

for (d_group, train), (d_group2, test) in zip(couples_X_Y_TR.items(), couples_X_Y_TE.items()):
    X_train, Y_train = train[0], train[1]    
    X_test, Y_test = test[0], test[1]     # we evaluate on dataset of same demographic group in January
    
    mod = RidgeCV(alphas=alpha_grid, fit_intercept=True, scoring='r2', cv=spl).fit(X_train, Y_train)  # 100-fold CV
    chosen = mod.alpha_
    tot_chosen.append(chosen)
    preds = lr.predict(X_test)

    r2 = r2_score(Y_test, preds)
    rmse = np.sqrt(mean_squared_error(Y_test, preds))
    
    print(f"\n\t \t DEMOGRAPHIC GROUP: {d_group}")
    print(f"Optimal parameter chosen: {np.around(chosen, 2)}")
    print(f"R2 score on January's sample: {r2: .5%}")
    print(f"RMSE score on January's sample: {rmse}")
    
    store_dem[d_group] = [r2, rmse]
    
print(f"\nThe avg. optimal Ridge parameter across demographic groups is {np.around(np.array(tot_chosen).mean(), 2)}")
```

### Elastic Net

```python
store_dem = {}

alpha_grid = np.concatenate((np.logspace(-4, 0, 50), np.arange(1.1, 1.5, 0.1)))    # weight of penalization (both)
l_grid = np.linspace(0.05, 1, 50).tolist()  # trade-of

spl = 100
tot_chosen_a = []
tot_chosen_l = []


for (d_group, train), (d_group2, test) in zip(couples_X_Y_TR.items(), couples_X_Y_TE.items()):
    X_train, Y_train = train[0], train[1]    
    X_test, Y_test = test[0], test[1]     # we evaluate on dataset of same demographic group in January
    
    mod = ElasticNetCV(l1_ratio=l_grid, alphas=alpha_grid, cv=spl, selection='cyclic', tol=1e-5, max_iter=int(1e4))  # 100-fold CV
    lr = mod.fit(X_train, Y_train.ravel())
    chosen_a = mod.alpha_
    chosen_l = mod.l1_ratio_
    tot_chosen_a.append(chosen_a)
    tot_chosen_l.append(chosen_l)
    preds = lr.predict(X_test)

    r2 = r2_score(Y_test, preds)
    rmse = np.sqrt(mean_squared_error(Y_test, preds))
    
    print(f"\n\t \t DEMOGRAPHIC GROUP: {d_group}")
    print(f"Optimal alpha chosen: {np.around(chosen_a, 5)}")
    print(f"Optimal lambda chosen: {np.around(chosen_l, 5)}")
    print(f"R2 score on January's sample: {r2: .5%}")
    print(f"RMSE score on January's sample: {rmse}")
    
    store_dem[d_group] = [r2, rmse]
    
print(f"\nThe avg. optimal penalization parameter across demographic groups is {np.around(np.array(tot_chosen_a).mean(), 2)}")
print(f"\nThe avg. optimal trade-off parameter across demographic groups is {np.around(np.array(tot_chosen_l).mean(), 2)}")
```

### Random Forest with Transfer Learning

```python
def create_features(X_train, X_test, Y_train, Y_test):
    ''' Function used to create engineered features as outputs of a deep neural network'''
    
    clear_session()
    
    model=Sequential()  #name
    model.add(BatchNormalization(input_shape=(X_train.shape[1],)))
    model.add(Dropout(0.2))
    model.add(Dense(30, name='first_hidden', kernel_constraint=MaxNorm(3.0)))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(Dense(100, name='second_hidden', kernel_constraint=MaxNorm(3.0)))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    model.add(Dense(30, name='third_hidden', kernel_constraint=MaxNorm(3.0)))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='linear', name='output'))
    model.compile(optimizer=RMSprop(momentum=0.9), loss=MeanSquaredError(), metrics=[RSquare(), RootMeanSquaredError()])  # loss_weights = ..., 
    model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=100, verbose=0, batch_size=32)
    #model.summary()
    
    wanted_layer = 'second_hidden'
    feature_model = Model(inputs=model.input, outputs=model.get_layer(wanted_layer).output)
    
    addit_feature_train = feature_model.predict(X_train)  # dropout not used when predictions, it knows in test mode automatically
    addit_feature_test = feature_model.predict(X_test)
    X_train_eng = np.concatenate((X_train, addit_feature_train), axis=1)
    X_test_eng = np.concatenate((X_test, addit_feature_test), axis=1)
    #print(f"Engineered X_train with number of features = {X_train_eng.shape[1]}")
    #print(f"Engineered X_test with number of features = {X_test_eng.shape[1]}")
    return X_train_eng, X_test_eng
```

```python
store_dem = {}

parameters={"n_estimators":[20,50, 100,150],
            "max_depth" : [1,3,5,8,15],
            "min_samples_leaf":[1,2,8],
           "max_features":[1.0,"sqrt"],
           "max_leaf_nodes": [None,2,10,20,50] }


for (d_group, train), (d_group2, test) in zip(couples_X_Y_TR.items(), couples_X_Y_TE.items()):
    X_train, Y_train = train[0], train[1]    
    X_test, Y_test = test[0], test[1]     # we evaluate on dataset of same demographic group in January
    
    X_train_eng, X_test_eng = create_features(X_train, X_test, Y_train, Y_test)
    
    mod = RandomForestRegressor(bootstrap=True, verbose=0, max_samples=0.8)
    grid = GridSearchCV(estimator=mod, cv=5, param_grid=parameters, scoring='r2', n_jobs=-1, verbose=1)  
    lr = grid.fit(X_train, Y_train.ravel())  
    preds = lr.predict(X_test)
    
    r2 = r2_score(Y_test, preds)
    rmse = np.sqrt(mean_squared_error(Y_test, preds))
    
    print(f"\n\t \t DEMOGRAPHIC GROUP: {d_group}")
    print(f"R2 score on January's sample: {r2: .5%}")
    print(f"RMSE score on January's sample: {rmse}")
    print(f"Best parameters: {lr.best_params_}")
    
    store_dem[d_group] = [r2, rmse]

```
